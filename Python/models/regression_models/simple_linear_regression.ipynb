# Simple Linear Regression

## Data pre-processing

Import libraries

```python
import numpy as np
import pandas as pd
# Matplotlib for ploting data
import matplotlib.pyplot as plt
# Sklearn (split the dataset)
from sklearn.model_selection import train_test_split
# Sklearn (lineal regression model)
from sklearn.linear_model import LinearRegression
```

Import dataset and split in testing and training set

```python
dataset = pd.read_csv('../../data/Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3)
```

## Build model

Create a simple lineal regression model with training set

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Evaluation of results

Predic with testing set

```python
y_pred = model.predict(X_test)
```

Results of trained model

```python
results = pd.DataFrame({
    'Data Test': y_test,
    'Prediction': np.around(y_pred, decimals=2) 
})

display(results)
```

## Data visualization

Plot training data

```python
plt.style.use('fivethirtyeight')
plt.scatter(X_train, y_train)
plt.plot(X_train, model.predict(X_train), color="orangered")
plt.title("Training - Salary based on experience")
plt.xlabel("Years of experience")
plt.ylabel("Salary ($)")
plt.show()
```

Plot testing data

```python
plt.style.use('fivethirtyeight')
plt.scatter(X_test, y_test)
plt.plot(X_train, model.predict(X_train), color="orangered")
plt.title("Testing - Salary based on experience")
plt.xlabel("Years of experience")
plt.ylabel("Salary ($)")
plt.show()
```
