# Support Vector Regression (SVR)

## Data pre-processing

Import libraries

```python
import numpy as np
import pandas as pd
# Matplotlib for ploting data
import matplotlib.pyplot as plt
# Scikit-learn (Standardize features)
from sklearn.preprocessing import StandardScaler
# Scikit-learn (lineal regression model)
from sklearn.linear_model import LinearRegression
# scikit-learn (svr features)
from sklearn.svm import SVR
```

Import dataset and split in dependent and independent variables

```python
dataset = pd.read_csv('../../../data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

Standardize idependent and dependent variable

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.reshape(-1, 1))
```

## Build model

Create a support vector regression model

```python
model = SVR(kernel = 'rbf')
model.fit(X, y.ravel())
```

## Data visualization

Revert standardized scaling of variables for create a axis for ploting data

```python
X_plot = sc_X.inverse_transform(X[:9,])
y_plot = sc_y.inverse_transform(y[:9,])
```

Plot Support Vector Regresion (SVR) Model

```python
plt.style.use('fivethirtyeight')
plt.scatter(X_plot, y_plot)
plt.plot(X_plot, sc_y.inverse_transform(model.predict(sc_X.transform(X_plot)).reshape(-1,1)), color ="orangered")
plt.title("Support Vector Regression")
plt.xlabel("Employee's position")
plt.ylabel("Salary ($)")
plt.show()
```
