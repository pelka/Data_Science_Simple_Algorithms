# Desicion Tree Regression

## Data pre-processing

Import libraries

```python
import numpy as np
import pandas as pd
# Matplotlib for ploting data
import matplotlib.pyplot as plt
# Scikit-learn (Split the dataset)
from sklearn.model_selection import train_test_split
# Scikit-learn (Standardize features)
from sklearn.preprocessing import StandardScaler
# Scikit-learn (Desicion Tree Regression)
from sklearn.tree import DecisionTreeRegressor
```

Import dataset and split in testing and training set

```python
dataset = pd.read_csv('../../data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

## Build model

Create a support vector regression model

```python
model = DecisionTreeRegressor(random_state = 0)
model.fit(X, y)
```

## Data visualization

Create a grid for ploting data

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

Plot Polynomial Regression Model

```python
plt.style.use('fivethirtyeight')
plt.scatter(X, y)
plt.plot(X_grid, model.predict(X_grid), color ="orangered")
plt.title("Polynomial Regression Model")
plt.xlabel("Employee's position")
plt.ylabel("Salary ($)")
plt.ticklabel_format(style='plain')
plt.show()
```
