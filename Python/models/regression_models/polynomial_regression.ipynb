# Polynomial Regression

## Data pre-processing

Import libraries

```python
import numpy as np
import pandas as pd
# Matplotlib for ploting data
import matplotlib.pyplot as plt
# Scikit-learn (lineal regression model)
from sklearn.linear_model import LinearRegression
# scikit-learn (polynomial features)
from sklearn.preprocessing import PolynomialFeatures
```

Import dataset and split in dependent and independent variables

```python
dataset = pd.read_csv('../../data/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

## Build model

Create a lineal regression model with dependent variable

```python
lin_model = LinearRegression()
lin_model.fit(X, y)
```

Fit the linear regression model with polynomial regression

```python
poly_reg = PolynomialFeatures(degree = 3)
x_poly = poly_reg.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(x_poly, y)
```

## Data visualization

### Comparation of lineal vs polynomial regressions

Plot Linear Regression Model

```python
plt.style.use('fivethirtyeight')
plt.scatter(X, y)
plt.plot(X, lin_model.predict(X), color ="orangered")
plt.title("Linear Regression model")
plt.xlabel("Employee's position")
plt.ylabel("Salary ($)")
plt.ticklabel_format(style='plain')
plt.show()
```

Create a grid for ploting data

```python
X_grid = np.arange(min(X), max(X), 0.5)
X_grid = X_grid.reshape(len(X_grid), 1)
```

Plot Polynomial Regression Model

```python
plt.style.use('fivethirtyeight')
plt.scatter(X, y)
plt.plot(X_grid, poly_model.predict(poly_reg.fit_transform(X_grid)), color ="orangered")
plt.title("Polynomial Regression Model")
plt.xlabel("Employee's position")
plt.ylabel("Salary ($)")
plt.ticklabel_format(style='plain')
plt.show()
```
